```python
df['BuildingArea'] = df['BuildingArea'].astype(int)
df['Landsize'] = df['Landsize'].astype(int)
df['BuildingArea'] = df['BuildingArea'].astype(float)
df['Landsize'] = df['Landsize'].astype(float)
```

## Re-sample the dataset

```python
# First, separate the dataframes based on the type
df_h = df[df['Type'] == 'h']
df_u = df[df['Type'] == 'u']
df_t = df[df['Type'] == 't']

# Get the number of 'h' type
num_h = len(df_h)

# Resample 'u' and 't' types
df_u_resampled = df_u.sample(n=num_h, replace=True, random_state=42)
df_t_resampled = df_t.sample(n=num_h, replace=True, random_state=42)

# Concatenate all dataframes
df1 = pd.concat([df_h, df_u_resampled, df_t_resampled])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=10)
```

```python
gradient_model = GradientBoostingRegressor(learning_rate=0.1, n_estimators=200)
gradient_model.fit(X_train, y_train)
gradient_model.score(X_test, y_test)
```

```python
rf_model = RandomForestRegressor(n_estimators=30)
rf_model.fit(X_train, y_train)
rf_model.score(X_test, y_test)
```

```python
import xgboost as xgb

# Data set for storing xgboost results (matrices)
dtrain_reg = xgb.DMatrix(X_train, y_train)
dtest_reg = xgb.DMatrix(X_test, y_test)

params = {"objective": "reg:squarederror", "tree_method": "hist", "device": "cuda"}  # Set the device to cuda

n = 100
xgboost_model = xgb.train(
    params=params,
    dtrain=dtrain_reg,
    num_boost_round=n,
)

# XGBoost's train function returns a Booster object and doesn't have a score method.
# You can use the predict method and then calculate the score using a metric function from sklearn.
y_pred = xgboost_model.predict(dtest_reg)

# Calculate the score using a metric function from sklearn, for example r2_score
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
print(score)
```

## Testing models

### Metrics

#### R2 score

```python
from sklearn.metrics import r2_score

gradient_model_score  = r2_score(y_test, gradient_model.predict(X_test))
print("Gradient R2:", gradient_model_score)

rf_model_score  = r2_score(y_test, rf_model.predict(X_test))
print("Random forest R2:", rf_model_score)

y_pred = xgboost_model.predict(dtest_reg)
xgboost_model_score  = r2_score(y_test, y_pred)
print("XGBoost R2:", xgboost_model_score)
```

#### MAE

```python
from sklearn.metrics import mean_absolute_error

# Compute the MAE for the gradient model
gradient_model_mae = mean_absolute_error(y_test, gradient_model.predict(X_test))
print("Gradient MAE:", gradient_model_mae)

# Compute the MAE for the random forest model
rf_model_mae = mean_absolute_error(y_test, rf_model.predict(X_test))
print("Random Forest MAE:", rf_model_mae)

# Compute the MAE for the XGBoost model
xgboost_model_mae = mean_absolute_error(y_test, y_pred)  # y_pred was computed in the previous step
print("XGBoost MAE:", xgboost_model_mae)
```

#### RMSE

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Compute the RMSE for the gradient model
gradient_model_rmse = np.sqrt(mean_squared_error(y_test, gradient_model.predict(X_test)))
print("Gradient RMSE:", gradient_model_rmse)

# Compute the RMSE for the random forest model
rf_model_rmse = np.sqrt(mean_squared_error(y_test, rf_model.predict(X_test)))
print("Random Forest RMSE:", rf_model_rmse)

# Compute the RMSE for the XGBoost model
xgboost_model_rmse = np.sqrt(mean_squared_error(y_test, y_pred))  # y_pred was computed in the previous step
print("XGBoost RMSE:", xgboost_model_rmse)
```

```python
def predict_price_xg(location, house_type, distance, bedroom2, bathroom, car, landsize, buildingArea, yearBuilt, lattitude, longtitude):
    # Initialize x as a DataFrame with the same columns as the training data
    x = pd.DataFrame(columns=X.columns)

    # Set the values in x that correspond to the features of the house you want to predict the price for
    x.loc[0] = [0] * len(X.columns)
    x['Type'] = house_type
    x['Distance'] = distance
    x['Bedroom2'] = bedroom2
    x['Bathroom'] = bathroom
    x['Car'] = car
    x['Landsize'] = landsize
    x['BuildingArea'] = buildingArea
    x['YearBuilt'] = yearBuilt
    x['Lattitude'] = lattitude
    x['Longtitude'] = longtitude
    if location in X.columns:
        x[location] = 1

    # Convert the DataFrame to a DMatrix
    x_dmatrix = xgb.DMatrix(x)

    return xgboost_model.predict(x_dmatrix)[0]
```

```python
def predict_price_rf(location,house_type,distance,bedroom2,bathroom,car,landsize,buildingArea,yearBuilt,lattitude,longtitude):    
    loc_index = np.where(X_test.columns==location)[0][0]

    x = np.zeros(len(X_test.columns))
    x[0] = house_type
    x[1] = distance
    x[2] = bedroom2
    x[3] = bathroom
    x[4] = car
    x[5] = landsize
    x[6] = buildingArea
    x[7] = yearBuilt
    x[8] = lattitude
    x[9] = longtitude
    if loc_index >= 0:
        x[loc_index] = 1

    return rf_model.predict([x])[0]
```
